Dataset of an E-learning platform, trying to find patterns between marks based on Gender, Parental Educational Qualification, level of preparatory courses completed etc to find if the pattern varies based on preparatory courses only or the marks are related  with other factors as well. Primary agenda is to find importance of and offer more preparatory courses before actually onboarding students for regular courses.

```python
# correlation
cor1 = d1.corr()
# heat Map using Seaborn 
sns.heatmap(cor1,xticklabels=cor1.columns,yticklabels=cor1.columns,annot=True)
```

High positive correlation seen amongst the numeric Variables, which shows there are strong relations between the Marks

#Data Visualisation with Insights

```python
sns.relplot(x='math score',y='reading score', hue = 'gender', data =d1 )
```

Region with high score in reading in low score in maths is more dominated with females, while region with high score in math and low in reading is more dominated with Males. Also the region with high/low in both shows subjects show a mixed trend for both genders

```python
sns.relplot(x='math score',y='reading score', hue = 'test preparation course', data =d1 )
```

Students who completed none of the preparation course are the one's getting least score, and students who have completed all preparation courses are getting higher scores in both. 

```python
sns.relplot(x='math score',y='reading score', hue = 'parental level of education', data =d1 )
```

Marks of students show no significant relation with their parental education.  

```python
# Calculating  Mean Score for male
mean_score_male = np.mean(d1.query("gender=='male'"))
mean_score_male
```

```python
# Calculating  Mean Score for female
mean_score_female = np.mean(d1.query("gender=='female'"))
mean_score_female 
```

Females have Higher Mean Scores for Reading and Writing but males have a higher mean score with maths.
